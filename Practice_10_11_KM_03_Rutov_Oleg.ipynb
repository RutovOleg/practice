Розрахуйте коефієнти розкладу біному Ньютона за допомогою генератора (https://uk.wikipedia.org/wiki/%D0%91%D1%96%D0%BD%D0%BE%D0%BC_%D0%9D%D1%8C%D1%8E%D1%82%D0%BE%D0%BD%D0%B0).

1. Степінь многочлена користувач має вводити з клавіатури (невід'ємне ціле число).

2. Програма має підраховувати біноміальні коефіцієнти для біномів усіх степенів, починаючи від 0 і закінчуючи тим, який ввів користувач.

3. Коефіцієнти мають виводитись у вигляді трикутника Паскаля.

Наприклад, якщо користувач введе число 5, програма має вивести результат:

1

1 1

1 2 1

1 3 3 1

1 4 6 4 1

1 5 10 10 5 1


```
def binomial_coefficients(power):
    yield 1
    if power > 0:
        if power > 1:

            def multiply(multiply_start_num, multiply_end_num):
                result = 1
                for num in range(multiply_start_num, multiply_end_num + 1):
                    result *= num
                return result

            for k in range(1, power):
                yield round(multiply(max(k, power - k) + 1, power) / multiply(1, min(k, power - k)))

        yield 1


# Функція для введення степеня біному
def input_degree():
    while True:                                                           # Цикл буде повторюватия, поки не буде введено число
        input_string = input('Please, input degree of myonial: ?')        # Введення значення
        try:                                                              # Переевірка, чи є введене значення числом
            input_value = int(input_string)
        except ValueError:                                                # Обробка помилки, якщо введено не число
            print('Degree must be an integer.\nPlease, try again')
        else:
            if input_value in range(0, 101):                              # Перевірка числового значення
                return input_value                                        # Вивдення значення з функції
            else:
                 print('Degree must be between 0 and 100.\nPlease, try again')  # Виведення повідомлення про помилку


myonial_degree = input_degree()

for current_power in range(0, myonial_degree + 1):
    for coefficient in binomial_coefficients(current_power):
        print(coefficient, end=' ')
    print()
```

6. Замість функції з п. 4 створіть генератор, який буде визначати послідовність книжкових сторінок для кожного зошита. Ці рядки можна потім вивести або у циклі, або за допомогою функції next(). Функція-декоратор має працювати з цим генератором так само, як і зі звичайною функцією.

```
def request_user_choice(info_message, *choice_variants):
    '''
    :param info_message: information text, will be print once before start of input
    :param choice_variants: value pairs: (selection code - string lowercase value, return value)
    :return: value depending on user choice
    '''
    print(info_message)
    while True:
        users_choice_code = input('Enter Your choice: ?').lower()
        for current_variant in choice_variants:
            if users_choice_code == current_variant[0]:
                return current_variant[1]
        print('Incorrect choice code. Please, try again\n')


def input_book_length(required_divider):
    error_message_template = '\nIncorrect book length.\n{}\nPlease, try again or enter "b" to exit program'
    print('\nPlease, input number of pages in book:'
          '\n* Input number must be an integer.'
          '\n* Input number must be between 1 and 1280.'
          f'\n* Input number must be multiple of {required_divider}')
    while True:
        input_string = input('\nEnter book length: ?')
        if input_string.lower() == 'b':
            raise exit(0)
        try:
            book_length = int(input_string)
        except ValueError:
            print(error_message_template.format('Book pages number must be an integer.'))
        else:
            if book_length in range(1, 1281):
                if book_length % required_divider == 0:
                    return book_length
                else:
                    possible_book_length = (book_length // required_divider + 1) * required_divider
                    if possible_book_length <= 1280:
                        print('\nIncorrect book length.'
                              f'\nBook pages number must be multiple of {required_divider}.'
                              f'\n{book_length} is not multiple of {required_divider}.'
                              f'\nBut you can change Your book length to {possible_book_length} pages.')
                        if request_user_choice('Are you want to do it (y/n)?', ('y', True), ('n', False)):
                            print(f'Great, book pages number is changed to {possible_book_length} pages.')
                            return possible_book_length
                        else:
                            print('So, input new book length or enter "b" to exit program')
                    else:
                        print(error_message_template.format(f'Book pages number must be multiple of {required_divider}.'))
            else:
                print(error_message_template.format('Book pages number must be between 1 and 1280.'))


def book_pages_display(displaying_mode):
    def decorator(generator):
        def return_function(*args, **kwargs):
            if displaying_mode:
                notebooks = []
                for notebook_pages_list in generator(*args, **kwargs):
                    notebooks.append(notebook_pages_list)
                return notebooks
            else:
                return generator(*args, **kwargs)
        return return_function
    return decorator


notebook_pages_number = request_user_choice(
'''
Please, select type of using notebooks:
* Type A: 16 pages (8  sheets / 4 big sheets)
    # Enter "a" or "A" to select this type.
* Type B: 24 pages (12 sheets / 6 big sheets)
    # Enter "b" or "B" to select this type.
* Type C: 32 pages (16 sheets / 8 big sheets)
    # Enter "c" or "C" to select this type.
''',
('a', 16), ('b', 24), ('c', 32))

book_pages_number = input_book_length(notebook_pages_number)

create_general_list = request_user_choice(
'''
Please, select decorator function mode:
whether to return all values once and in general list?
* Yes
    # Enter "1" to select this answer
* No
    # Enter "0" to select this answer
''',
('0', False), ('1', True))


@book_pages_display(create_general_list)
def book_pages(book_length, notebook_length):
    notebook_half_length = notebook_length // 2
    for current_first_page in range(1, book_length, notebook_length):
        notebook_pages_list = []
        current_last_page = current_first_page + notebook_length - 1
        for i in range(0, notebook_half_length, 2):
            notebook_pages_list.append(current_last_page - i)
            notebook_pages_list.append(current_first_page + i)
            notebook_pages_list.append(current_first_page + i + 1)
            notebook_pages_list.append(current_last_page - i - 1)
        yield notebook_pages_list


print(f'Number of needed notebooks: {book_pages_number // notebook_pages_number}')

for print_notebook in book_pages(book_pages_number, notebook_pages_number):
    print(print_notebook)
```
