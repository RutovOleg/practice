### **Завдання 1 (1,5 бали)**

У 1939 році друком вийшов роман американського письменика Ернеста Вінсента Райта "Ґедсбі". Особливу увагу публіки він здобув через те, що письменник жодного разу не використав у романі літеру "e" (роман складається з більше ніж 50,000 слів), яка є найбільш вживаною літерою в англійській мові.

Порахуйте кількість входжень кожної літери у текст роману. Виведіть список літер англійського алфавіту, а також інформацію про те, скільки відсотків від усіх літер тексту складає кожна літера алфавіту.

Зауваження:

1. Програма має ігнорувати будь-які символи, окрім літер.

2. Переведіть усі літери у верхній або у нижній регістр.

3. Під час виведення відсортуйте літери у порядку зменшення їхнього відсотка. Літера "e" має стояти на останньому місці у цьому списку.

4. Виведіть перші 5 та останні 5 літер відсортованого списку разом з відсотковими значеннями, що їм відповідають.

```
from string import ascii_lowercase as alphabet


counting_dict = {letter: 0 for letter in alphabet}
alphabet_letters = list(alphabet)
letters_number = 0

with open('gadsby.txt', 'r') as text_file:
    for line in text_file:
        for symbol in line.lower():
            if symbol in alphabet_letters:
                counting_dict[symbol] += 1
                letters_number += 1

fractions_dict = {key: round(counting_dict[key] / letters_number * 100, 3) for key in alphabet_letters}

while True:
    order_not_changed = True
    for i in range(0, len(alphabet_letters) - 1):
        if counting_dict[alphabet_letters[i]] < counting_dict[alphabet_letters[i + 1]]:
            buffer = alphabet_letters[i]
            alphabet_letters[i] = alphabet_letters[i + 1]
            alphabet_letters[i + 1] = buffer
            order_not_changed = False
    if order_not_changed:
        break


def print_heading(heading):
    print('-' * 21)
    print(heading)
    print('-' * 21)

print('English alphabet:', alphabet)
print('\nEnglish | Percentage'
      '\nalhabet |     in'
      '\nletters |    text')

print_heading('Most frequent letters')
for i in range(0, 5):
    print(f'   {alphabet_letters[i]}    |   {fractions_dict[alphabet_letters[i]]}%')

print_heading('  The rarest letters')
for i in range(-5, 0):
    print(f'   {alphabet_letters[i]}    |   {fractions_dict[alphabet_letters[i]]}%')

print_heading('    Other letters')
for key in alphabet_letters[5:-5]:
    print(f'   {key}    |   {fractions_dict[key]}%')
```

### **Завдання 2 (1,5 бали)**

Створіть CSV-файл, що буде містити від 5 до 10 пісень (творів) вашого улюбленого музиканта. Назва файлу - ім'я музиканта (композитора) або назва гурту.

1. Обов'язково додайте до файлу рядок з назвами двох колонок (Song, Year) або (Composition, Year).

2. Після запису файлу роздрукуйте назву файлу та усі наявні у файлі рядки (включно із заголовками).

3. Для запису та зчитування рядків скористайтесь функціями csv.DictWriter() та csv.DictReader().

```
import csv


#Seting work data
compositor_name = 'Sabaton'
compositions = [
    {'Song': 'The Last Stand',
     'Year': '2016'},
    {'Song': 'Ghost Division',
     'Year': '2008'},
    {'Song': 'The attack of the Dead Men',
     'Year': '2019'},
    {'Song': 'Bismarck',
     'Year': '2019'},
    {'Song': 'The Red Baron',
     'Year': '2019'},
    {'Song': 'The Lost Battalion',
     'Year': '2016'},
    {'Song': 'Shiroyama',
     'Year': '2016'}
]

file_name = f'{compositor_name}.csv'

#Writing file
with open(file_name, 'w', newline='') as writing_file:
    file_writer = csv.DictWriter(writing_file, fieldnames=['Song', 'Year'])
    file_writer.writeheader()
    for composition in compositions:
        file_writer.writerow(composition)

#Reading file
print(f'Reading file "{file_name}"...\n')
with open(file_name, newline='') as reading_file:
    file_reader = csv.DictReader(reading_file)
    headings = file_reader.fieldnames
    print(headings[0], '   |   ', headings[1])
    print('-'*40)
    for composition in file_reader:
        print(composition[headings[0]], ' | ', composition[headings[1]])
        print('-'*40)
```

```
import json


def display_progress(complete_message, *progress_messages):
    for message in progress_messages:
        print(message)
        yield
        print(complete_message)
        yield


#Setting work data
searching_image = '000000000001.jpg'
file_name = 'image_info_test-dev2017.json'

progressor = display_progress('Complete.',
                              'Loading COCO Database...',
                              f'Searching for "{searching_image}" image...')

next(progressor)
with open(file_name) as coco_dataset:
    data = json.load(coco_dataset)
next(progressor)

next(progressor)
for image in data['images']:
    if image['file_name'] == '000000000001.jpg':
        break
next(progressor)

print('\nNumber of images is', len(data['images']),
      '\nNumber of categories is', len(data['categories']))

print(f'\nSearched "{searching_image}" image parameters:'
      f'\n"coco_url" = "{image["coco_url"]}"'
      f'\n"id" = {image["id"]}'
      f'\n"height" = {image["height"]}'
      f'\n"width" = {image["width"]}')
```
