### **Завдання 2 (1,5 бали)**

Написати програму підрахунку коренів квадратного рівняння без попередніх перевірок.

1. Користувач має вводити коефіцієнти a, b та c з клавіатури.

2. Програма має обробляти помилки ділення на 0, кореня від'ємного числа та інші, а також виводити відповідні повідомлення у разі виникнення таких помилок.

3. Якщо рівняння не має коренів, программа має генерувати помилку за допомогою ключового слова raise, а потім обробляти її.

4. Правильно написана програма при виникненні різних помилок має виводити текст помилки, а також різні повідомлення про те, в чому суть цих помилок. Необроблених помилок бути не повинно.

```
import math

a = 0
b = 0
c = 0
ans = set()
print('ax^2 + bx + c = 0')
try:
    a = float(input('a =?'))
    b = float(input('b =?'))
    c = float(input('c =?'))
    try:
        d = math.sqrt(b**2 - 4*a*c)
        ans.add(round((0 - b + d) / (2*a), 3))
        ans.add(round((0 - b - d) / (2*a), 3))
    except ValueError as sqrtValueError:
        print(f"Помилка!\n{sqrtValueError}\n"
              "Взяття квадратного кореня з від'ємного числа. Дискримінант менше 0.")
        raise Warning(f'Рівняння {a}x^2 + {b}x + {c} = 0 не має коренів')
    except ZeroDivisionError as incAZeroDivisionError:
        print(f"Помилка!\n{incAZeroDivisionError}\n"
              "Спроба ділення на 0. Неприпустиме значення a (a = 0).")
except ValueError as convertValueError:
    print(f"Помилка!\n{convertValueError}\n"
          "Введене значення коефіцієнта не є числом.")
except Warning as InternalWarning:
    print(InternalWarning)
# Вловлювання непередбачуваних помилок:
except Exception as UnknownError:
    print(f'Помилка!\n{UnknownError}\nНевідома помилка.')
else:
    print(f'Множина корнів рвняння {a}x^2 + {b}x + {c} = 0'
          f'\nдорівнює {ans}'
          f'\n*значення коренів округлені до 3-х знаків після точки')
```
