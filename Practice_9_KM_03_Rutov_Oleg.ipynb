### **Завдання 1 (1 бал)**

Продавець роздрібного товару влаштовує розпродаж. Для всіх товарів знижка однакова - 60%. Щоб вивісити нові цінники, продавець хоче скласти таблицю, в якій для кожного товару буде підрахована його нова вартість та сума знижки.

Написати програму, яка буде генерувати таку таблицю. Вартість одиниць товарів до розпродажу подано у списку products_list. 

1. Для підрахунку невідомих значень повинні використовуватись вбудовані функції вищих порядків.

2. Вартість з урахуванням знижки та суми знижок для кожного товару повинні зберігатись у двох окремих списках. Порядок товарів у нових списках повинен бути такий самий, як і в списку products_list.

3. Кожне підраховане числове значення має бути округлене до сотих.

4. Дані повинні виводитись у вигляді:

**Discount table:**

**29.25 11.7 17.55**

**48.99 19.6 29.39**

**...**


products_list = [[29.25, 48.99, 99.98, 124.65, 214.30, 543.90, 799.85]]


```
products_list = [29.25, 48.99, 99.98, 124.65, 214.30, 543.90, 799.85]

new_prices = list(map(lambda price: round(price * 0.4, 2), products_list))
discounts = list(map(lambda new_price, old_price: round(old_price - new_price, 2), new_prices, products_list))

print('Discount table')
for i in range(0, len(products_list)):
    print(products_list[i], new_prices[i], discounts[i])
```

```
# Додаткова версія програми:
# * Уведення початкових цін та відсотку знижки
#   відбувається вручну користувачем з клавіатури.
#   Введення організоване за допомогою фцнкції вищих порядків.


# Функція для введення дробових чисел з перевірками
def input_float(string_check_condition,  # Функція перевірки уведеного рядка
                 value_check_condition,   # Функція перевірки уведеного значення
                 info_message, input_message, error_message):
    '''
    :param string_check_condition: input string checking function: if it will return True,
    function will return entered string;
    :param value_check_condition: converted value checking function: if it will return True,
    function will return converted value;
    :param info_message: message, will be shown before each input;
    :param input_message: text, will be filled into input line;
    :param error_message: message, will be shown when ia error is detected;
    :return: input string or already converted to number input value.
    '''
    print(info_message)
    while True:                                                   # Цикл буде повторюватия, поки не буде введено число
        input_string = input(input_message)                       # Введення значення
        if string_check_condition(input_string):                  # Перевірка введеного рядка
            return input_string.lower()                           # Повернення уведеного рядка при виконанні умови
        else:
            try:                                                  # Переевірка, чи є введене значення числом
                input_value = float(input_string)
            except ValueError:                                    # Обробка помилки, якщо введено не число
                print(error_message, '\nPlease, try again')
            else:
                if value_check_condition(input_value):            # Перевірка числового значення переданою функцією
                    return input_value                            # Вивдення значення з функції
                else:
                    print(error_message, '\nPlease, try again')   # Виведення повідомлення про помилку


# Введення відсотку знижки:
discount_coefficient = 1 - 0.01 * input_float(lambda input_string: False,
                                              lambda input_value: input_value > 0 and input_value < 100,
                                              'Please, input discount (%):',
                                              '? ',
                                              'Discount must be between 0 and 100.')

# Створення та заповнення введенеми з клавіатури значеннями списку products_list:
products_list = []
print('Input prices of products:')
while True:
    input_price = input_float(lambda input_string: input_string.lower() == 'end',
                              lambda input_value: input_value > 0,
                              '  Please, input price of product:\n  (type "end" to end prices input)',
                              '  ? ',
                              'Price must be a positive number.')
    if input_price != 'end':
        products_list.append(round(input_price, 2))
    else:
        break


# Основна частина програми:
new_prices = list(map(lambda price: round(price * discount_coefficient, 2), products_list))
discount_sums = list(map(lambda new_price, old_price: round(old_price - new_price, 2), new_prices, products_list))
# Виведення таблиці знижок:
print('\nDiscount table')
for i in range(0, len(products_list)):
    print(products_list[i], new_prices[i], discount_sums[i])
```

### **Завдання 2 (2 бали)**

Зазвичай рік має 365 днів. Однак у високосні роки додається ще один день - 29 лютого. Правила визначення високосного року такі:

- Будь-який рік, що ділиться на 400 - високосний.

- З тих років, що залишились, будь-який рік, що ділиться на 100 - **не** високосний.

- З тих років, що залишились, будь-який рік, що ділиться на 4 - високосний.

- Усі інші роки, що залишились - **не** високосні.

1) Створіть функцію, в якій за допомогою вбудованих функцій вищих порядків будуть визначатися високосні роки серед перелічених років у списку years. Створена функція має повертати список високосних років.

2) Створіть функцію вищих порядків, яка прийматиме на вхід попередню функцію та визначатиме кількість днів у місяці. Також функція повинна приймати на вхід ще 2 параметри: ціле число від 1 до 12 (місяць) та рік, як ціле чотирицифрове число. Ці значення корисувач має вводити з клавіатури. Перевірку введених даних слід виконувати поза межами функцій.

```
# Роз'яснення до розв'язку:
#   Наскільки я зрозумів, користувач може ввести лише рік від 1990 до 2020,
#   оскільки в іншому разі цього року не буде в списку years, а значить,
#   ми не зможемо перевірити його на високосніть методом, описаним в умові.


import numpy as np

years = np.arange(1900, 2020+1, 1)


def get_leap_years():
    return filter(lambda year: year % 400 == 0 or (year % 100 != 0 and year % 4 == 0), years)


def get_days_number(leap_years_func, year, month):
    if month == 2:
        if year in leap_years_func():
            day_number = 29
        else:
            day_number = 28
    elif month in {4, 6, 9, 10}:
        day_number = 30
    else:
        day_number = 31

    return day_number


# Функція для введення цілих чисел з перевірками
def input_int(value_check_func, input_message, error_message):
    '''
    :param value_check_func: converted value checking function: if it will return True, 
    function will return converted value;
    :param input_message: text, will be filled into input line;
    :param error_message: message, will be shown when ia error is detected;
    :return: converted to number input value.
    '''
    while True:                                               # Цикл буде повторюватия, поки не буде введено число
        input_string = input(input_message)                   # Введення значення
        try:                                                  # Переевірка, чи є введене значення числом
            input_value = int(input_string)
        except ValueError:                                    # Обробка помилки, якщо введено не число
            print(error_message, '\nPlease, try again')
        else:
            if value_check_func(input_value):                 # Перевірка числового значення переданою функцією
                return input_value                            # Вивдення значення з функції
            else:
                print(error_message, '\nPlease, try again')   # Виведення повідомлення про помилку


print('This month has',
      get_days_number(get_leap_years,
                      input_int(lambda year_num: year_num in range(1900, 2021),
                                'Please, input a number of year: ?',
                                'Year number must be an integer ranging from 1900 to 2020'),
                      input_int(lambda month_num: month_num in range(1, 13),
                                'Please, input a number of month: ?',
                                'Number of month must be an integer ranging from 1 to 12')),
      'days.')
```
